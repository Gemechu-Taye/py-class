```python
# Experiment 2 (B) Recursive process

def fact(n):
    if(n==1):
        return 1
    else:
        return n*fact(n-1)
n = int(input("Enter a number: "))
if(n==0):
    print("Factorial of ",n," = 1")
elif(n<0):
    print("Factorial of a negative number is not defined.")
else:
    fact(n)
    print("Factorial of ",n," = ",fact(n))
```

```python
# Experiment 2 (B) Fibonacci Series of a number

def fib(n):
    if(n==1):
        return 0
    elif(n==2):
        return 1
    else:
        return fib(n-1)+fib(n-2)
n = int(input("Enter the number of terms required in Fibonacci sequence: "))
n1 = 0
if(n==0): 
    exit
elif(n==1):
    print("Fibonacci sequence: ",n1)
else:
    print("Fibonacci Sequence: ",end=" ")
    for i in range(1,n+1):
        print(fib(i),end=" ")
```

```python
# Experiment 4 (A)

def linearsearch(a,se): 
    count=0 
    list=[] 
    for i in range(len(a)): 
        if(a[i]==se): 
            count=count+1 
            list.append(i) 
    if(count==0): 
            print("The search element is not found in the list.") 
    else: 
            print("The search element is present at index:",list) 
a=input("Enter the list of elements: ").split(" ") 
se=input("Enter the search element:") 
linearsearch(a,se)
```

```python
# Experiment 4 (A) - Binary Search

def binarysearch(a,beg,end,se): 
    if (end>=beg): 
        mid=(beg+end)//2 
        if(a[mid]==se): 
            print("The search element is present at index:",mid) 
        elif(a[mid]<se): 
            binarysearch(a,mid+1,end,se) 
        else: 
            binarysearch(a,beg,mid-1,se) 
    else: 
        print("The search element is not found in the list.") 
a=[int(x) for x in input("Enter the list of elements in sorted "
" order without repeating the elements: ").split(" ")] 
se=int(input("Enter the search element:")) 
binarysearch(a,0,len(a)-1,se)
```

```python
# Experiment 4 (B) Bubble Sort 

def bubblesort(a): 
    for i in range(len(a)-1): 
        for j in range(len(a)-i-1): 
            if(a[j]>a[j+1]): 
                a[j],a[j+1]=a[j+1],a[j] 
a=[int(x) for x in input("Enter the list of elements: ") 
    .split(" ")] 
bubblesort(a) 
print("The sorted list:",a)
```

```python
# Experiment 4 (B) Selection Sort 

def selectionsort(a): 
    for i in range(len(a)-1): 
        smallest=i 
        for j in range(i+1,len(a)): 
            if(a[j]<a[smallest]): 
                smallest=j 
    a[i],a[smallest]=a[smallest],a[i] 
a=[int(x) for x in input("Enter the list of elements: ") 
    .split(" ")] 
selectionsort(a) 
print("The sorted list:",a)
```

```python
# Experiment 4 (B) Insertion Sort

def insertionsort(a): 
    for i in range(1,len(a)): 
        temp=a[i] 
        j=i-1 
        while(j>= 0 and temp<a[j]): 
            a[j+1] = a[j] 
            j=j-1 
            a[j+1]=temp 
a=[int(x) for x in input("Enter the list of elements: ") 
    .split(" ")] 
insertionsort(a) 
print("The sorted list:",a)
```
